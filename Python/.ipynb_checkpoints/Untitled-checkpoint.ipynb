```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
engine = create_engine('mysql+pymysql://ucfnama:dixavuvoyu@dev.spatialdatacapture.org:3306/ucfnama')
conn = engine.raw_connection()
```

```python
data31254 = pd.read_csv('../Austria/31254.csv')
data31255 = pd.read_csv('../Austria/31255.csv')
data31256 = pd.read_csv('../Austria/31256.csv')
dataTirol = pd.read_csv('../Austria/Tirol.csv')
dataVienna = pd.read_csv('../Austria/Vienna.csv')
```

```python
alldata = data31254.append(data31255).append(data31256).append(dataTirol).append(dataVienna)
```

```python
alldata.shape
```

```python
austria = alldata.groupby(['STREET','POSTCODE','CITY'])['LON','LAT'].mean()
```

```python
dataVienna.plot(kind="scatter", x="LON", y="LAT", alpha=0.4)
plt.show()
```

```python
austria.to_sql('sbs_austria',engine,if_exists = "replace")
```
