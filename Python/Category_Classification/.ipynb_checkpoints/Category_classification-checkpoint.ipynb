```python
# Loop through the category column and append the category name to a list! 

categories = new_df['category']
category_single = []
for category in categories:
    # Check if the category is already in the list. If yes, go to the next one!
    # Take only the 3 first words of the strings! 
    first_words = category.split()[0:3]
    category_clean = " ".join(first_words)
   # print(category_clean)
    if category_clean in category_single:
        continue
    else: # Category not in list, append it to the list
        category_single.append(category_clean)
```

```python
for i in category_single:
    #print(i)
    if i.startswith('Commune'):
        need_replace = i 
        only_one = need_replace.split()[0]
      #  print(only_one)
        category_single = [test.replace(need_replace,only_one) for test in category_single]
    elif i.startswith('Localité'):
        need_replace = i 
        only_one = need_replace.split()[0]
      #  print(only_one)
        category_single = [test.replace(need_replace,only_one) for test in category_single]
    elif i.startswith('Ville'):
        need_replace = i 
        only_one = need_replace.split()[0]
      #  print(only_one)
        category_single = [test.replace(need_replace,only_one) for test in category_single]
    elif i.startswith('État'):
        need_replace = i 
        only_one = need_replace.split()[0]
      #  print(only_one)
        category_single = [test.replace(need_replace,only_one) for test in category_single]
```
